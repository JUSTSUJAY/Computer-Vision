# Steps carried out during Canny Edge Detection

1. Filters out noise using gaussian Blur
2. Finds strengt and direction of edges using sobel filters
3. Applies non maximum Suppression to isolate strongest edges and thin them to one pixel wide lines
4. Uses Hysteresis to isolate the best edges


# How does hysteresis isolation works?
Hysteresis isolation has two threshold         1. Low 2. High

above high threshold - strong edges - kept

below low threshold - weak edges - discarded

and pixels that fall between high and low threshold are either kept if it is connected to a strong edge or discarded otherwise

```python
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
import numpy as np
```

```python
def Canny_Edge_detect(image):
    img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    wide = cv2.Canny(img,30,60)
    tight = cv2.Canny(img,120,240)
    plt.subplot(121)
    plt.imshow(wide,cmap='gray')
    plt.subplot(122)
    plt.imshow(tight,cmap='gray')
    
```

```python
img = cv2.imread('images/sunflower.jpg')
Canny_Edge_detect(img)
```
