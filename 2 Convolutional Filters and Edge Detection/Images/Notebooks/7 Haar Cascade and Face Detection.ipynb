 these classifiers in the OpenCV library and use feature-based classification cascades that learn to isolate and detect faces in an image. You can read the [original paper proposing this approach here.](https://www.cs.cmu.edu/~efros/courses/LBMV07/Papers/viola-cvpr-01.pdf)

# Lib Import

```python
# import required libraries for this section
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# Face Detector Function

```python
def face_detected(p,scale_factor,min_neighbors):
    # load in cascade classifier
    face_cascade = cv2.CascadeClassifier('detector_architectures/haarcascade_frontalface_default.xml')

    rgb = cv2.cvtColor(p,cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(p,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray,scale_factor,min_neighbors)
    print ('We found ' + str(len(faces)) + ' faces in this image')
    print ("Their coordinates and lengths/widths are as follows")
    print ('=============================')
    print (faces)
    img_with_detections = np.copy(rgb)   # make a copy of the original image to plot rectangle detections ontop of

    # loop over our detections and draw their corresponding boxes on top of our original image
    for (x,y,w,h) in faces:
        # draw next detection as a red rectangle on top of the original image.  
        #Argument 4 - R(255,0,0)G(0,255,0)B(0,0,255)
        # Argument - 5 thickness of the rectangular lines
        cv2.rectangle(img_with_detections,(x,y),(x+w,y+h),(0,255,0),5)  

    # display the result
    plt.figure(figsize=(20,10))
    plt.imshow(img_with_detections)
```

# Image Read

```python
# load in color image for face detection
image = cv2.imread('images/multi_faces.jpg')
group_photo = cv2.imread('images/group photo.jpg')
```

`detectMultiScale`(`image`,`scaleFactors`,`minNeighbors`)
more faces for smaller `scaleFactor`and lower value for `minNeighbors`, but raising these values often produces better matches. Modify these values depending on your input image.

```python
face_detected(image,4,6)
```

```python
face_detected(group_photo,2,3)
```
