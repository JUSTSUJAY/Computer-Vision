In the previous notebook we used sobel filter for edge detection, but the problem with high pass filter is that 
if directly used onto an image they leave a lot of noise in the image and it becomes difficult to extract the information from the image

Hence it becomes important to use a low pass filter before using a high class filter 

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('images/brain_MR.jpg')
brain = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(brain,cmap='gray')
```

# Gaussian Blur

```python
blur = cv2.GaussianBlur(brain,(9,9),0)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('original gray')
ax1.imshow(brain, cmap='gray')

ax2.set_title('blurred image')
ax2.imshow(blur, cmap='gray')
```

```python
# High-pass filter 

# 3x3 sobel filters for edge detection
sobel_x = np.array([[ -1, 0, 1], 
                   [ -2, 0, 2], 
                   [ -1, 0, 1]])


sobel_y = np.array([[ -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])

filtered_brain_x = cv2.filter2D(blur,-1,sobel_x)
filtered_brain_y = cv2.filter2D(blur,-1,sobel_y)

# f, (ax1, ax2,ax3, ax4, ax5, ax6) = plt.subplots(1, 3, figsize=(20,10))
f, (ax1, ax2,ax3,ax4) = plt.subplots(1, 4, figsize=(20,10))


ax1.set_title('original Brain')
ax1.imshow(brain, cmap='gray')

ax2.set_title('Blurred Brain')
ax2.imshow(blur, cmap='gray')

ax3.set_title('Filtered Brain X')
ax3.imshow(filtered_brain_x, cmap='gray')


ax4.set_title('Filtered Brain X')
ax4.imshow(filtered_brain_y, cmap='gray')
```
