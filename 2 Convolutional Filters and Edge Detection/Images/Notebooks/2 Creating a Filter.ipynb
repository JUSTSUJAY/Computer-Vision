```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image = cv2.imread('images/city_hall.jpg')
img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Convert image into grayscale
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

# Creating a Kernel

```python
'''Sobel Filter - Used to detect verticle edges
format -> [-k,0,k]
detect abrupt intensity changes in certain direction'''
sobel_x = np.array([[-1,0,1],
                  [-5,0,5],
                  [-1,0,1]])
filtered_x = cv2.filter2D(gray,-1,sobel_x) #represents that do not change the format of output image and keep it same as input image
plt.imshow(filtered_x,cmap='gray')
```

# Creating a Binary Image

```python
#  Set a threshold, so that points above threshold will be white/visible
retval, binary_image = cv2.threshold(filtered_x,50,255,cv2.THRESH_BINARY)
plt.imshow(binary_image,cmap='gray')
```

When I set threshold to 50 we got more noise in the image. because the grass is detected
To reduce the noise, increase the threshold

```python
#  Set a threshold, so that points above threshold will be white/visible
retval, binary_image = cv2.threshold(filtered_x,150,255,cv2.THRESH_BINARY)
plt.imshow(binary_image,cmap='gray')
```

```python
#  Set a threshold, so that points above threshold will be white/visible
retval, binary_image = cv2.threshold(filtered_x,200,255,cv2.THRESH_BINARY)
plt.imshow(binary_image,cmap='gray')
```

```python
sobel_y = np.array([[-1,-5,-1],
                   [0,0,0],
                   [1,5,1]])
filtered_y = cv2.filter2D(gray,-1,sobel_y)
plt.imshow(filtered_y,cmap='gray')
```

```python
retval,binary_image = cv2.threshold(filtered_y,150,255,cv2.THRESH_BINARY)
plt.imshow(binary_image,cmap='gray')
```

# Reminder

sobel_x detects vertical lines nd sobel_y detects horizontal line

Relation is inverse

```python
curved_lane = cv2.imread('images/curved_lane.jpg')
lane = cv2.cvtColor(curved_lane,cv2.COLOR_BGR2GRAY)
plt.imshow(lane,cmap='gray')
```

```python
filtered_lane_x = cv2.filter2D(lane,-1,sobel_x)
filtered_lane_y = cv2.filter2D(lane,-1,sobel_y)
ret_val,binary_image_1 = cv2.threshold(filtered_lane_x,100,255,cv2.THRESH_BINARY)
plt.imshow(binary_image_1,cmap='gray')
```

```python
ret_val,binary_image_2 = cv2.threshold(filtered_lane_y,100,255,cv2.THRESH_BINARY)
plt.imshow(binary_image_2,cmap='gray')
```
