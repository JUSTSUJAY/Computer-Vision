```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
%matplotlib inline
```

```python
image = cv2.imread('images/rainbow_flag.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
image.shape
```

```python
level1 = cv2.pyrDown(image)
level2 = cv2.pyrDown(level1)
level3 = cv2.pyrDown(level2)

f,(ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(20,10))

ax1.set_title('Original')
ax1.imshow(image)

ax2.imshow(level1)
ax2.set_xlim(0,image.shape[0])
ax2.set_ylim(0,image.shape[1])

ax3.imshow(level2)
ax3.set_xlim(0,image.shape[0])
ax3.set_ylim(0,image.shape[1])

ax4.imshow(level3)
ax4.set_xlim(0,image.shape[0])
ax4.set_ylim(0,image.shape[1])
```
