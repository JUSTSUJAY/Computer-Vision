```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('pizza_bluescreen.jpg')
print('This image is:', type(image), 
      ' with dimensions:', image.shape)
```

```python
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
# Defining a section boundary
lower_blue = np.array([0,0,230])
upper_blue = np.array([50,70,255])
```

```python
# Creating a mask
mask = cv2.inRange(image_copy,lower_blue,upper_blue)
plt.imshow(mask,cmap='gray')
```

```python
# Show Pizza without blue bg
masked_pizza = np.copy(image_copy)
masked_pizza[mask!=0] = [0,0,0]
plt.imshow(masked_pizza)
```

```python
# Let us change the banground from black to something else
bg = cv2.imread('anime-girl-watching-sunset-fence-4k-0s.jpg')
cropped_bg = bg[:514,:816]
cropped_bg[mask==0] = [0,0,0]
plt.imshow(cropped_bg)
```

```python
final = cropped_bg+masked_pizza
plt.imshow(final)
```

# Image 2 - Car with green BG

```python
car = cv2.imread('car_green_screen.jpg')
# CV2 initially loads the image in BGR Format which we need to convert into RGB format
car = cv2.cvtColor(car,cv2.COLOR_BGR2RGB)
plt.imshow(car)
```

```python
# Next step is to declare the boundaries 
lower_green = np.array([0,230,0])
upper_green = np.array([100,255,100])
```

```python
car.shape
```

```python
# Creating a mask
masked = cv2.inRange(car,lower_green,upper_green)
plt.imshow(masked,cmap='gray')
```

```python
background = cv2.imread('anime-girl-watching-sunset-fence-4k-0s.jpg')
plt.imshow(background)
```

```python
masked_car = np.copy(car)
masked_car[masked!=0] = 0
plt.imshow(masked_car)
```

```python
cropped_background = background[:450,:660]
cropped_background[masked==0] = 0
plt.imshow(cropped_background)
```

```python
final = cropped_background+masked_car
plt.imshow(final)
```
