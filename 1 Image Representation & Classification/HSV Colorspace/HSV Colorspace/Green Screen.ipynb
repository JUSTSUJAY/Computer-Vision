```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
img = cv2.imread('car_green_screen-Copy1.jpg')
car = np.copy(img)
car = cv2.cvtColor(car,cv2.COLOR_BGR2RGB)
plt.imshow(car)
```

```python
# Color Channels
r = car[:,:,0]
g = car[:,:,1]
b = car[:,:,2]

f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(30,10))

ax1.set_title('Red')
ax1.imshow(r,cmap='gray')

ax2.set_title('Green')
ax2.imshow(g,cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b,cmap='gray')

plt.tight_layout()
plt.show()
```

```python
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
```

```python
# Color Channels
h = car[:,:,0]
s = car[:,:,1]
v = car[:,:,2]

f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(30,10))

ax1.set_title('hue')
ax1.imshow(h,cmap='gray')

ax2.set_title('saturation')
ax2.imshow(s,cmap='gray')

ax3.set_title('value')
ax3.imshow(v,cmap='gray')

plt.tight_layout()
plt.show()
```

```python
# Seperation Channels
# RGB
lower_rgb = np.array([0,180,0])
upper_rgb = np.array([200,255,200])
# HSV
lower_hsv = np.array([150,0,0])
upper_hsv = np.array([180,255,255])
```

```python
# RGB Mask
rgb_mask = cv2.inRange(car,lower_rgb,upper_rgb)
rgb_cp = np.copy(car)
rgb_cp[rgb_mask!=0] = [0,0,0]
plt.imshow(rgb_cp)
```

```python
# Hsv Mask
hsv_mask = cv2.inRange(car,lower_rgb,upper_rgb)
hsv_cp = np.copy(car)
hsv_cp[hsv_mask!=0] = [0,0,0]
plt.imshow(hsv_cp)
```
