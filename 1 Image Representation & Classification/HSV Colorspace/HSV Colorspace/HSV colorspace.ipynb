```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
%matplotlib inline
balloons = cv2.imread('water_balloons.jpg')
```

```python
plt.imshow(balloons)
```

```python
img_cp = cv2.cvtColor(balloons,cv2.COLOR_BGR2RGB)
plt.imshow(img_cp)
```

```python
# RGB Channels
img_cp[:1]
# R -> 116,116,117 -> index = 0
# G -> 81,81,80 -> index = 1
# B -> 88,88,88 -> index = 2
# Hence
```

```python
img_cp[:5][:,:,0]
```

```python
r = img_cp[:,:,0]
g = img_cp[:,:,1]
b = img_cp[:,:,2]
```

```python
f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(30,10))

ax1.set_title('Red')
ax1.imshow(r,cmap='gray')

ax2.set_title('Green')
ax2.imshow(g,cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b,cmap='gray')

plt.tight_layout()
plt.show()
```

```python
# Things are not clear from RGB channel so we will try using HSV colorspace
balloons_hsv = cv2.cvtColor(balloons,cv2.COLOR_BGR2HSV)
plt.imshow(balloons_hsv)
```

```python
# HSV Channels
h = balloons_hsv[:,:,0]
s = balloons_hsv[:,:,1]
v = balloons_hsv[:,:,2]
```

```python
f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(30,10))

ax1.set_title('HUE')
ax1.imshow(h,cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s,cmap='gray')

ax3.set_title('Value')
ax3.imshow(v,cmap='gray')

plt.tight_layout()
plt.show()
```

```python
# Hue is pretty high for pink color in first image

# Seperation for rgb channels
lower_rgb = np.array([180,50,50])
upper_rgb = np.array([255,255,230])

# Seperation for hue
lower_hue = np.array([160,0,0])
upper_hue = np.array([180,255,255])
```

```python
# Masks = RGB
mask_rgb = cv2.inRange(img_cp,lower_rgb,upper_rgb)
masked_img = np.copy(img_cp)
masked_img[mask_rgb==0] = [0,0,0]
plt.imshow(masked_img)
```

```python
# Masks - HUE
mask_hue = cv2.inRange(balloons_hsv,lower_hue,upper_hue)
masked_img = np.copy(img_cp)
masked_img[mask_hue==0] = [0,0,0]
plt.imshow(masked_img)
```
